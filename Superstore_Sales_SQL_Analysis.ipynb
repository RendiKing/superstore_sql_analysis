Superstore Sales SQL Analysis

```python
import pandas as pd
import sqlite3
from google.colab import files
import zipfile
import io
```

```python
uploaded = files.upload()
```

```python
# Unzip the uploaded file
zip_file = zipfile.ZipFile(io.BytesIO(uploaded['Sample - Superstore.csv.zip']))
zip_file.extractall()
```

```python
# Load the extracted CSV
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

```

```python
conn = sqlite3.connect('superstore.db')
df.to_sql('superstore', conn, if_exists='replace', index=False)
```

Query 1: Total Sales by Region

```python
query = '''
SELECT Region, SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC;
'''

pd.read_sql_query(query, conn)
```

Query 2: Top 10 Customers by Total Spend

```python
query = '''
SELECT "Customer Name", SUM(Sales) AS Total_Spend
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Spend DESC
LIMIT 10;
'''

pd.read_sql_query(query, conn)
```

Number of Orders by State

```python
query = '''
SELECT State, COUNT("Order ID") AS Num_Orders
FROM superstore
GROUP BY State
ORDER BY Num_Orders DESC;
'''

pd.read_sql_query(query, conn)
```

Average Profit per Category

```python
query = '''
SELECT Category, AVG(Profit) AS Avg_Profit
FROM superstore
GROUP BY Category
ORDER BY Avg_Profit DESC;
'''

pd.read_sql_query(query, conn)
```

Monthly Sales Trend

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
query = '''
SELECT strftime('%Y-%m', "Order Date") AS Month, SUM(Sales) AS Monthly_Sales
FROM superstore
GROUP BY Month
ORDER BY Month;
'''

pd.read_sql_query(query, conn)
```
